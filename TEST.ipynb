```
import tensorflow as tf
from tensorflow import keras

new_model = tf.keras.models.load_model(Data_dir+'my_model.h5')

# Check its architecture
new_model.summary()
```

```
import pickle
# loading
with open(Data_dir+'tokenizer.pickle', 'rb') as handle:
    tokenise = pickle.load(handle)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

max_sequence_len=88

def make_lyrics(seed_text, next_words):
    pred_index=[]
    for i in range(next_words):
        token_list = tokenise.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list],
                     maxlen=max_sequence_len-1,padding='pre')
        #print(token_list.shape)
        token_list = np.reshape(token_list, (1, max_sequence_len-1, 1))
        predicted = new_model.predict(token_list, verbose=0)
        predicted_index =  np.argmax(predicted)
        pred_index.append(predicted_index)
        


        #predicted_index=1
        output_word = ""
        for word, index in tokenise.word_index.items():
            if index == predicted_index:
                output_word = word
                break
        seed_text += " " + output_word
    print(seed_text)
    return seed_text
```

```
make_lyrics('Do you',5)
```

```
line1 = make_lyrics("The club isn't the best place",8)
```
