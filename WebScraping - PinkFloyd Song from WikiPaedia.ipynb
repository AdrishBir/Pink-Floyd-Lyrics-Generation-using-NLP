```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import bs4 as bs
import urllib.request
```

```
from google.colab import drive
drive.mount('/content/gdrive',force_remount=False)
Data_dir = '/content/gdrive/My Drive/Colab_Notebooks/Datasets/Pink Floyd/'
```

```
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_songs_recorded_by_Pink_Floyd#Songs').read()
soup = bs.BeautifulSoup(source,'html')
row,title = [],[]
table = soup.find('table',class_='wikitable sortable plainrowheaders')
table_rows = table.find_all('tr')
for tr in table_rows:
    th = tr.find_all('th')
    td = tr.find_all('td')
    row.extend([i.text for i in td])
    title.extend([i.text for i in th])    
```

```
writer, lead_vocal, album, year = [],[],[],[]
for i in range(0,(len(row)-4),5):
  writer.append(row[i])
  lead_vocal.append(row[i+1])
  album.append(row[i+2])
  year.append(row[i+3])
```

```
writer = [i.replace('\n','') for i in writer]
album = [i.replace('\n','') for i in album]
year = [i.replace('\n','') for i in year]
lead_vocal = [i.replace('\n','') for i in lead_vocal]
```

```
title= []
for tr in table_rows:
    th = tr.find_all('th')
    title.extend([i.text for i in th])
    
title = title[6:]
title = [i.replace('\n','') for i in title]
```

```
details = { 
    'title' : title, 
    'writer' : writer, 
    'lead' : lead_vocal,
    'album': album,
    'year': year 
} 
wiki_df = pd.DataFrame(details)
wiki_df.head()
```
